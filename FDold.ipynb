```python
# find latitude relation

plt.figure(figsize=(20, 10))

lat_rad = np.zeros(90)
lat_num = np.zeros(90)


# sum radiation for one degree
for i in range(len(df)):
    if (df['H'][i]) > 0:
        lat_rad[int(round(df['lat'][i]))] += df['H'][i]
        lat_num[int(round(df['lat'][i]))] += 1
            
# average calculation
for i in range(90):
    lat_rad[i] /= lat_num[i]
            
plt.plot(lat_rad,'o')
    
pf_coef = np.polyfit(range(28,43),lat_rad[range(28,43)],1)

print pf_coef

p = (np.poly1d(pf_coef))

plt.plot(sy.linspace(0,70), p(sy.linspace(0,70)), 'r-')
```
