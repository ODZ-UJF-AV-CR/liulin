```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sy

from scipy.optimize import curve_fit

import pandas as pd  # data parsing library
from matplotlib import cm, colors

from glob import glob
```

```python
# convert time from LS data

df = pd.read_csv('./LSdataArchiv.csv', sep=',', header=0)
#df['time'] = datetime.datetime(1900,1,1)
for i in range(len(df)):
    df['time'][i] = datetime.datetime(df['year'][i],df['month'][i],df['day'][i],df['hour'][i]-1)
df.to_csv('./LSradiation.csv') 
```

```python
# find latitude relation

plt.figure(figsize=(20, 10))

lat_rad = np.zeros(90)
lat_num = np.zeros(90)

for filename in glob('./output/*.csv'): # search for 'csv' files
 
    # read file
    df = pd.read_csv(filename,sep=',', header=0 )
    records = len(df)  
    
    # sum radiation for one degree
    for i in range(records):
        if (df['0'][i]) > 16:
            lat_rad[int(round(df['lat'][i]))] += df['0'][i]
            lat_num[int(round(df['lat'][i]))] += 1
            
# average calculation
for i in range(90):
    lat_rad[i] /= lat_num[i]
            
plt.plot(lat_rad,'o')
    
pf_coef = np.polyfit(range(25,63),lat_rad[range(25,63)],1)

print pf_coef

p = (np.poly1d(pf_coef))

plt.plot(sy.linspace(25,63), p(sy.linspace(25,63)), 'r-')
```

```python
# subtract latitude ralation

df2 = pd.DataFrame(columns=['time', 'radiation'])

for filename in glob('./output/*.csv'): # search for 'csv' files
 
    # read file
    df = pd.read_csv(filename,sep=',', header=0 )
    records = len(df)  
    
    for i in range(records): 
        if (df['0'][i]) > 30:
            df2.loc[len(df2)+1] = [pd.to_datetime(df['time'][i]), df['0'][i] - p(df['lat'][i])]

    #break

df2 = df2.set_index('time')
df2 = df2.sort_index()

    
#plt.figure(figsize=(20, 10))
#plt.plot(df2['time'],df2['radiation'])
#df2.plot(figsize=(20,10), style='o')
df2.plot(figsize=(20,10), style='.')
#plt.gcf().autofmt_xdate()
df2.to_csv('./radiation.csv') # save merged data to output directory
```
